# EDA (EXploratory Data Analysis) or Data Visualisation

# Seaborn Visual Library
We'll try to illustrate some charts based on couple of columns throughout the dataset with the Seaborn library. Let's try something simple with only a single X axis values againts its own value counts.

```python
# Let's try to visualize the 'Segment' column from the dataset.

sns.set_style('whitegrid')
sns.countplot(x='Segment',data=Eda, palette='rainbow')
```

```python
# Let's try to visualize the 'Country' column from the dataset.

sns.set_style('whitegrid')
sns.countplot(x='State',data=Eda, palette='rainbow')
```

```python
# Let's try to visualize the 'Category' column from the dataset.

sns.set_style('whitegrid')
sns.countplot(x='Category',data=Eda, palette='rainbow')
```

```python
# Let's try to visualize the 'Category' column from the dataset.

sns.set_style('whitegrid')
sns.countplot(x='Sub-Category',data=Eda, palette='rainbow')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Region',data=Eda, palette='rainbow')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Ship Mode',data=Eda, palette='rainbow')
```

```python
# Let's try to visualize the 'Region' column from the Shipmode column dataset standpoints.

region_hist = sns.FacetGrid(Eda, col='Ship Mode', palette='rainbow')
region_hist.map(plt.hist, 'Region')
region_hist.set_ylabels('Number')
```

```python
# Let's try to visualize the 'Category' column from the Ship mode column dataset standpoints.

category_hist = sns.FacetGrid(Eda, col='Ship Mode', palette='rainbow')
category_hist.map(plt.hist, 'Category')
category_hist.set_ylabels('Number')
```

```python
plt.figure(figsize = (20,8))
sns.barplot(x='State', y='Profit', data = Eda)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.heatmap(Eda.corr(), annot=True, linecolor='white', linewidths=3)
plt.xticks(size=10)
plt.yticks(size=10)
plt.title('Variables-Correlation')
plt.savefig('Variables-Correlation.png')
plt.show()
```
